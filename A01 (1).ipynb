```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

## Bateria y tableta

```python
df= pd.read_csv("C:/Users/munoz/OneDrive/Escritorio/lab_apre_est/bateria_tableta.txt")
df.head()
```

### - Realiza regresiones polinomiales de orden 1, 3, 5 y 7.

```python
x1 = df.carga.values.reshape([-1,1])
y1= df.duracion
```

## Orden 1

```python
lr=LinearRegression()
lr.fit(x1,y1)
lr.intercept_
```

```python
lr.coef_
```

### - Compara el R2 que te da cada modelo.

## Orden 3

```python
x2=x1**2
x3= x1**3
X3=np.hstack([x1,x2,x3])
lr=LinearRegression()
lr.fit(X3,y1)
lr.intercept_
```

```python
lr.coef_
```

### - Compara el R2 que te da cada modelo.

```python
r2_score(y1,lr.predict(X3))
```

### - Crea una gráfica con una nube de dispersión (datos originales) y tus modelos.

Si gráficaramos nuestra gréfica como lo vimos en clase, nos daría algo así. Esto porque nuestra x1, que es como se unen nuestros puntos, no esta ordenada, por lo que, va saltando de un punto a otro sin ningún orden. Provocando que se vea de esta manera nuestra gráfica.

```python
plt.scatter(x1,y1)
plt.plot(x1,lr.predict(X3))
```

Por esta razon, decidí investigar sobre como puedo ordenar la forma en que se conectan los puntos y tener una línea contuina. Para esto utilizamos np.argsort(x1.ravel()), que nos dan los índices en los que vamos a ordenar nuestra variable de x1, el .ravel() nos ayuda a aplanar nuestro array en una sola dimensión. Luego x1s = x1[x], ya es nuestra nueva variable de x1 pero ordenada de menor a mayor gracias a los índices, después ordenamos nuestras predicciónes según x1s (variable ya ordenada correctamente) y esta nueva variable la llamamos ys1. Y por último, las gráficamos mediante plt.plot(x1s,y1s), y podemos observar una gráfica mucho más sencilla y ordenada para su comprensión.

```python
plt.scatter(x1,y1)
x = np.argsort(x1.ravel())
x1s= x1[x]
y1s = lr.predict(X3)[x]
plt.plot(x1s, y1s)
```
