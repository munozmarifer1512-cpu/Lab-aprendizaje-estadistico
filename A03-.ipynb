```python
df22=df2.drop('model',axis=1)
```

```python

x2=df22.cyl.values.reshape([-1,1])
x3=df22.disp.values.reshape([-1,1])
x4=df22.hp.values.reshape([-1,1])
x5=df22.drat.values.reshape([-1,1])
x6=df22.wt.values.reshape([-1,1])
x7=df22.qsec.values.reshape([-1,1])
x8=df22.vs.values.reshape([-1,1])
x9=df22.am.values.reshape([-1,1])
x10=df22.gear.values.reshape([-1,1])
x11=df22.carb.values.reshape([-1,1])
y1=df22.mpg
n1=len(y1)
unos=np.ones([n1,1])
X=np.hstack([unos,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y1, test_size=0.4, random_state=137)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(x_train)

import statsmodels.api as sm

xe = sm.add_constant(x_train)
ols = sm.OLS(y_train, xe)
results = ols.fit()
results=ols.fit()
results.summary()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(x_test)

import statsmodels.api as sm

xe = sm.add_constant(x_test)  # aade la columna de unos
ols = sm.OLS(y_test, xe)
results = ols.fit()
results=ols.fit()
results.summary()
```

```python
x_test_scaled=sm.add_constant(x_test)
y_pred=results.predict(x_test)
r2_score(y_test,y_pred)
```
