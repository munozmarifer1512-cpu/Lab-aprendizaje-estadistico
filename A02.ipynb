```python
x1=df2.CRIM.values.reshape([-1,1])
x2=df2.ZN.values.reshape([-1,1])
x3=df2.INDUS.values.reshape([-1,1])
x4=df2.CHAS.values.reshape([-1,1])
x5=df2.NOX.values.reshape([-1,1])
x6=df2.RM.values.reshape([-1,1])
x7=df2.AGE.values.reshape([-1,1])
x8=df2.DIS.values.reshape([-1,1])
x9=df2.RAD.values.reshape([-1,1])
x10=df2.TAX.values.reshape([-1,1])
x11=df2.PTRATIO.values.reshape([-1,1])
x12=df2.B.values.reshape([-1,1])
x13=df2.LSTAT.values.reshape([-1,1])
y1=df2.MEDV
n1=len(y1)
unos=np.ones([n1,1])
X=np.hstack([unos,x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13])
ols=sm.OLS(y1,X)
resultados=ols.fit()
resultados.summary()
```

```python
#CRIM
es1 = (x1 - np.min(x1)) / (np.max(x1 - np.min(x1)))
print("CRIM")
print("Min: ", es1.min(), "Max: ", es1.max())

```

```python
for col in df2.columns:
    if col=="MEDV" :
        break
    #Paso 1
    minval=df2[col].min()
    maxval=df2[col].max()
    #Paso 2
    aux=(df2[col]-minval)/(maxval-minval)
    df2[col]=aux
df2.describe()
    
```
